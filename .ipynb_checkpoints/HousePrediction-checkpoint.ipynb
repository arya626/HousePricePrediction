```python
def get_null_percentage(data, cols):
    null_perc = []
    for i in cols:
        if(data[i].isnull().sum()>0):
            null_perc.append([i,data[i].isnull().sum(), round((data[i].isnull().sum()/data.shape[0])*100,2)])

    null_perc_df = pd.DataFrame(null_perc, columns  = ['Columns','#_of_rows', 'Percentage'])
    null_perc_df.sort_values(by=['#_of_rows'], inplace=True, ascending = False)
    null_perc_df = null_perc_df.reset_index(drop = True)
    return null_perc_df
```

From the table, we can replace categorical values having "nan" with 0 and rest starting from 1. The intuition behind replacing "nan" with 0 defines that the particular attribute is not present for that house. Replacing it with mode misinterprets the target attribute. Ex: few houses has no Alley or Pool or Garage etc. these attributes for such property is replaced with 0.


Few other houses has missing value (where we are not sure if those attributes are present or not). Such attributes are replaced with mode of that column

```python
train_data1["Alley"].replace(list(train_data1['Alley'].unique()), [0,1,2], inplace = True)
train_data1['MasVnrType'].replace(list(train_data1['MasVnrType'].unique()), [1,0,2,3,0], inplace = True)
train_data1['BsmtQual'].replace(list(train_data1['BsmtQual'].unique()), [1,2,3,0,4], inplace = True)
train_data1['BsmtCond'].replace(list(train_data1['BsmtCond'].unique()), [1,2,0,3,4], inplace = True)
train_data1['BsmtExposure'].replace(list(train_data1['BsmtExposure'].unique()), [1,2,3,4,0], inplace = True)
train_data1['BsmtFinType1'].replace(list(train_data1['BsmtFinType1'].unique()), [1,2,3,4,5,0,6], inplace = True)
train_data1['BsmtFinType2'].replace(list(train_data1['BsmtFinType2'].unique()), [1,2,0,3,4,5,6], inplace = True)
train_data1['FireplaceQu'].replace(list(train_data1['FireplaceQu'].unique()), [0,1,2,3,4,5], inplace = True)
train_data1['GarageType'].replace(list(train_data1['GarageType'].unique()), [1,2,3,4,0,5,6], inplace = True)
train_data1['GarageFinish'].replace(list(train_data1['GarageFinish'].unique()), [1,2,3,0], inplace = True)
train_data1['GarageQual'].replace(list(train_data1['GarageQual'].unique()), [1,2,3,0,4,5], inplace = True)
train_data1['GarageCond'].replace(list(train_data1['GarageCond'].unique()), [1,2,0,3,4,5], inplace = True)
train_data1['PoolQC'].replace(list(train_data1['PoolQC'].unique()), [0,1,2,3], inplace = True)
train_data1['Fence'].replace(list(train_data1['Fence'].unique()), [0,1,2,3,4], inplace = True)
train_data1['MiscFeature'].replace(list(train_data1['MiscFeature'].unique()), [0,1,2,3,4], inplace = True)

train_data1['Electrical'].replace(np.nan, train_data1['Electrical'].mode()[0], inplace = True)
```
